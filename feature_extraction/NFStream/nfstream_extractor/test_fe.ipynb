```python
from streamer import NFStreamer
import warnings
import pandas as pd
import glob
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', 400)
pd.set_option('display.max_rows', 100)
```

```python
main_dir = "/home/abdelkader.elmahdaou/lustre/data_sec-um6p-st-sccs-6sevvl76uja/IDS/mahdaouy/test_pcaps"
```

```python
files = glob.glob(f'{main_dir}/*.pcap')
len(files)
```

```python
idle = 3 * 60
active = 30 * 60
```

```python
dfs = []
for file in files:
    print(file)
    my_streamer = NFStreamer(source=file, # or network interface
                             idle_timeout=idle,
                             active_timeout=active)
    dfi = my_streamer.to_pandas()
    dfs.append(dfi)
print("All done")
```

```python
df = pd.concat(dfs)
```
