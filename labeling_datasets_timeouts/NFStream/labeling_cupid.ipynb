```python
def read_dataset_data(out_dir, dataset):
    files = glob.glob(out_dir+"/temp/"+dataset+"/*.csv")
    dfs = []
    for file in files:
        name = file.split('/')[-1].split('.')[0]
        dfi = pd.read_csv(file)
        dfi['name'] = name
        dfs.append(dfi)
    df = pd.concat(dfs)
    return df
```
